# Data Processing

```
##import all fits files
hdu_list1 = fits.open("6V.fits")
hdu_list2 = fits.open("DC_10ms.fits")
hdu_list3 = fits.open("DC_10ms_cover_with.fits")
hdu_list4 = fits.open("DC_10ms_cover_without.fits")
```

```
##get all hdu 
image_data1 = hdu_list1[0].data
image_data2 = hdu_list2[0].data
image_data3 = hdu_list3[0].data
image_data4 = hdu_list4[0].data
```

```
##plot the original data with signal
print(type(image_data1))
print(image_data1.shape)
plt.imshow(image_data1, cmap="magma")
plt.colorbar()


```

```
##find the area of image with dark noise
plt.imshow(image_data1[:, 0:500], cmap="magma")
plt.colorbar()
```

```
## plot the first DC
print(type(image_data2))
print(image_data2.shape)
stats.describe(image_data2)
plt.imshow(image_data2, cmap="magma")
plt.colorbar()
```

```
##plot the second DC
print(type(image_data3))
print(image_data3.shape)
plt.imshow(image_data3, cmap="magma")
plt.colorbar()
```

```
##plot the third DC
print(type(image_data4))
print(image_data4.shape)
plt.imshow(image_data4, cmap="magma")
plt.colorbar()
```

```
## substruction of first DC from original data and plot the first results
image_data_old=image_data1 - image_data2
image_data_auf=image_data1 - image_data3
image_data_ohne=image_data1-image_data4
plt.imshow(image_data_old, cmap="magma")
plt.colorbar()
```

```
## plot the part without the signal and with substracted DC1
plt.imshow(image_data_old[:, 0:500], cmap="magma")
plt.colorbar()
```

```
##statistics of the part without signal and with substracted DC1
print('Min:', np.min(image_data_old[:, 0:500]))
print('Max:', np.max(image_data_old[:, 0:500]))
print('Mean:', np.mean(image_data_old[:, 0:500]))
print('Stdev:', np.std(image_data_old[:, 0:500]))
##print(np)
```

```
## plot the part without the signal and with substracted DC2
plt.imshow(image_data_auf[:, 0:500], cmap="magma")
plt.colorbar()
```

# Repeat part of statistics analysis with if loop

```
hdu_list1 = fits.open("1_9mACW_10ms_1.fits")
hdu_list2 = fits.open("DC_10ms.fits")
hdu_list3 = fits.open("DC_10ms_deckel_auf.fits")
hdu_list4 = fits.open("DC_10ms_ohne_deckel.fits")
```

```
image_data1 = hdu_list1[0].data
image_data2 = hdu_list2[0].data
image_data3 = hdu_list3[0].data
image_data4 = hdu_list4[0].data
list=(image_data2, image_data3, image_data4)
```

```

for x in list:
    image_data=image_data1 - x
 
    print('Min:', np.min(image_data[:, 0:500]))
    print('Max:', np.max(image_data[:, 0:500]))
    print('Mean:', np.mean(image_data[:, 0:500]))
    print('Stdev:', np.std(image_data[:, 0:500]))
```
